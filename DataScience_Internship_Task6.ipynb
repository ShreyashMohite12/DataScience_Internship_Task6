```python
# Problem 1 
def smart_parking(capacity, logs):
    parked = 0
    peak = 0
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        peak = max(peak, parked)
    
    status = "Available" if parked < capacity else "Full"
    alert = "Yes" if parked > capacity else "No"
    
    return {
        "Currently Parked Vehicles": parked,
        "Peak Usage": peak,
        "Parking Status": status,
        "Capacity Alert": alert
    }

# Example
print(smart_parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"]))
```

```python
# Problem 2
def delivery_time(distance, traffic, weather):
    base_time = distance * 5  
    
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    
    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20
    
    return f"Estimated Delivery Time: {base_time} minutes"

# Example
print(delivery_time(8, "High", "Rainy"))
```

```python
# Problem 3
def theatre_occupancy(total_seats, booked_seats):
    occupancy = (len(booked_seats) / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 80:
        status = "Almost Full - Consider Extra Show"
    else:
        status = "Seats Available"
    
    return {
        "Occupancy": f"{occupancy:.2f}%",
        "Show Status": status
    }

# Example
print(theatre_occupancy(200, [1]*150))
```

```python
# Problem 4
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)
    
    if avg < 50:
        status = "Normal"
    elif 50 <= avg < 80:
        status = "Warning"
    else:
        status = "Critical"
    
    return {
        "Average CPU Load": f"{avg:.2f}%",
        "Server Status": status
    }

# Example
print(server_load([45, 60, 70, 85, 90]))
```

```python
# Problem 5
def classroom_usage(resources):
    overused = [res for res, hrs in resources.items() if hrs > 8]
    alert = "Yes" if overused else "No"
    
    return {
        "Overused Resources": overused,
        "Energy Alert": alert
    }

# Example
print(classroom_usage({"Projector":6,"AC":9,"Lights":4}))
```

```python
# Problem 6
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    return {
        "Confirmed Registrations": confirmed,
        "Waitlisted Users": waitlisted,
        "Registration Status": status
    }

# Example
print(event_registration(100, 105))
```
